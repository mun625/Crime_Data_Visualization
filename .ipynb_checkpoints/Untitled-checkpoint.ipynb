# Crime Analytics : Visualization of Incident Reports

In this notebook, We will analyze the crime datas of SanFrancisco and Seattle. These datas are real crime data from Summer 2014 provided by Coursera Lecture course.

Initialize variable and define some functions to analyze crime data. I recommend you just skip this section.

```python
import sys
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
hours = list(range(24))
months = list(range(1,13))

def ConvertToInt(var):
    try:
        var = int(var)
    except:
        pass
    return var

def CsvToLists(fp):
    csv_file = open(fp)
    data_lists = []
    for line in csv_file:
        data_lists.append(line.rstrip('\n').split(','))
    return data_lists

def CategorizedByDay(data):
    day_dict = {}
    for day in days:
        day_dict[day] = []

    for item in data:
        if item[3] in days:
            day_dict[item[3]].append(item)

    return day_dict

def CategorizedByHour(data):
    hour_dict = {}
    for hour in hours:
        hour_dict[hour] = []

    for item in data:
        only_hour = ConvertToInt(item[5].split(':')[0])
        if only_hour in hours:
            hour_dict[only_hour].append(item)
    return hour_dict

def CategorizedByMonth(data):
    month_dict = {}
    for month in months:
        month_dict[month] = []

    for item in data:
        only_month = ConvertToInt(item[4].split('/')[0])
        if only_month in months:
            month_dict[only_month].append(item)
    return month_dict

def CategorizedByCategory(data):
    category_dic = {}
    for item in data:
        if item[1] != 'Category':
            category_dic[item[1]] = []

    for item in data:
        if item[1] in category_dic:
            category_dic[item[1]].append(item)
    return category_dic
```

Parse the crime data that we want to analyze (Sanfrancisco's and Seattle's)

```python
sf_data = CsvToLists("sanfrancisco_incidents_summer_2014.csv")
se_data = CsvToLists("seattle_incidents_summer_2014.csv")
```

## 1. Crime by Incident time

Anlazying crime by time. You can check the time that incidents occurs frequently.<br>
First will analyze by hour, and second will analyze by day.

```python
# First : will analyze by hour
sf_time = CategorizedByHour(sf_data)
# set axis array to plot the graph
time_x = np.array([])
time_y = np.array([])
for time in hours:
    time_x = np.append(time_x, time)
    time_y = np.append(time_y, len(sf_time[time]))
    
print "As you can see below graph1, at dawn(1am - 9am) incidents occurs less frequently."
# plot number vs time
plt.plot(time_x, time_y)
plt.xlabel("Hour")
plt.ylabel("Number (of crime)")
plt.axhline(y=len(sf_time[9]), color='r')
plt.title('Figure1. Number(of crime) vs Hour')
plt.show()


# Second : will analyze by day
sf_day = CategorizedByDay(sf_data)
# set axis array to plot the graph
day_x = np.array([])
day_y = np.array([])
x = np.array(range(7))
for day in days:
    day_x = np.append(day_x, day)
    day_y = np.append(day_y, len(sf_day[day]))
    
print "Seeing below graph2, on weekend crimes occur more frequently compared to weekday"
# plot number vs time
plt.xticks(x, day_x)
plt.plot(x, day_y)
plt.xlabel("DayOfWeek")
plt.ylabel("Number (of crime)")
plt.axhline(y=(len(sf_day['Thursday'])+len(sf_day['Friday']))/2, color='r')
plt.title('Figure2. Number(of crime) vs Day')
plt.show()
```

## 2. Crime category

There are many types of incidents. This time, I will show you types of crimes and analyze each category data.

### First, Let's check what type of criminal occured.

```python
sf_category = CategorizedByCategory(sf_data)
category_number_list = []
for key, value in sf_category.iteritems():
    category_number_list.append([key, len(value)])
    
# Sort category by frequency(how many times it happened)
def SortListByNum(my_list):
    if len(my_list) == 0:
        return []
    sorted_list = []
    while(True):
        num = 0
        idx = 0
        for i in range(len(my_list)):
            if int(my_list[i][1]) > num:
                num = int(my_list[i][1])
                idx = i
        sorted_list.append(my_list[idx])
        del my_list[idx]
        if len(my_list) == 0:
            break
    return sorted_list
    
print "If you see below graph2, some category crime occur very frequently"
sorted_list = SortListByNum(category_number_list)
x = np.array(range(len(sorted_list)))
x_axis = np.array([])
y_axis = np.array([])
for item in sorted_list:
    x_axis = np.append(x_axis, item[0])
    y_axis = np.append(y_axis, item[1])
# plot number vs category
plt.figure(figsize=(16,6))
ax = plt.subplot(111)
ax.bar(x, y_axis, width=1)
plt.xlabel("Category Type")
plt.ylabel("Number (of crime)")
plt.title('Figure3. Number(of crime) vs Category')
plt.show()

print "Here are informations above first 4 categories"
print "--------------------Top 4 Category (most frequently occured)--------------------"
for item in sorted_list[0:4]:
    print item[0], item[1]
print "\nFrom now We will analyze the above 4 types of incidents"
```

### From now, we will analyze the above 4 types of incidents
Let's analyze "LARCENY/THEFT", "OTHER OFFENSES", "NON-CRIMINAL", "ASSAULT"

```python
sf_theft = sf_category["LARCENY/THEFT"]
sf_offense = sf_category["OTHER OFFENSES"]
sf_noncriminal = sf_category["NON-CRIMINAL"]
sf_assault = sf_category["ASSAULT"]

def PlotByHour(my_data):
    sf_time = CategorizedByHour(my_data)
    time_x = np.array([])
    time_y = np.array([])
    for time in hours:
        time_x = np.append(time_x, time)
        time_y = np.append(time_y, len(sf_time[time]))
    
    plt.plot(time_x, time_y)
    plt.xlabel("Hour")
    plt.ylabel("Number (of crime)")
    plt.title('Figure1. Number(of crime) vs Hour')
    plt.show()

def PlotByDay(my_data, title_name):
    sf_day = CategorizedByDay(my_data)
    day_x = np.array([])
    day_y = np.array([])
    x = np.array(range(7))
    for day in days:
        day_x = np.append(day_x, day)
        day_y = np.append(day_y, len(sf_day[day]))
    
    plt.xticks(x, day_x)
    plt.plot(x, day_y)
    plt.xlabel("DayOfWeek")
    plt.ylabel("Number (of crime)")
    plt.title(title_name)
    plt.show()
    
print "Below 4 graphs are explain the frequencies of each categories by day"
PlotByDay(sf_theft, 'Figure4. Number(of THEFT) vs Day')
PlotByDay(sf_offense, 'Figure5. Number(of OFFENSES) vs Day')
PlotByDay(sf_noncriminal, 'Figure6. Number(of NON-CRIMINALS) vs Day')
PlotByDay(sf_assault, 'Figure7. Number(of ASSAULTS) vs Day')
```

Let's compare above graphs to Figure2 (which explains about general crime v.s. day of week). <br>
General crimes occur more frequently at week day(Monday to Friday). <br>
<b>However, the OFFENSE incidents shows more frequently on weekday(see Fig5.)</b>
